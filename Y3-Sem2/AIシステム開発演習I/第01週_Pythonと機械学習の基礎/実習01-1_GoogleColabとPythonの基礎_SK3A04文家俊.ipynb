### 2-4. 複合データ型
リスト、タプル、辞書について違いを整理し、必要であればコードブロックを作成して確認する。
* リスト（list）：複数の変数を並べたもの
```
xlist = [1, 2, 3, 4, 5]
```
* タプル（tuple）：定義後に中の要素を変更できない
```
xtuple = (1, 2, 3)
```
* 辞書（dictionary）：キーと値をセットで格納
```
xdict = {"apple":1, "orange":2, "banana":3}
```

````
# 特に、リストの代入や変更、スライスなどを試そう。コードブロックを追加してもよい。
xlist = [1, 2, 3, 4, 5]
print(xlist[0]) # 最初の要素
print(xlist[-1]) # 最後の要素
print(xlist[1:3]) # 1番目から2番目までの要素を取り出す
print(xlist)

xlist2 = (1, 2, 3, 4, 5)

xlist2[0] = 1000
print(xlist2)
````

### 2-5. 制御構文の基本
*  if、forなどの行末は「:」（コロン）
*  ブロック内かどうかはインデントで判断（中括弧は使用しない）
```
if x > 0:
    print("x is positive.")
elif x < 0:
    print("x is negative.")
else:
    print("x is zero.")
```
* for文は「for 変数 in ～:」の形式（リスト、タプル、辞書の中身を順番に取り出したりもできる）
```
for i in range(5):
    print(i)
```





````
# xの値を変えてif～elif～else文を試す
x = 0
if x > 0:
  print("x is positive.")
elif x < 0:
  print("x is negative.")
else:
  print("x is zero.")
````

````
# for文を試す
for x in range(10):
  print(x)
````

