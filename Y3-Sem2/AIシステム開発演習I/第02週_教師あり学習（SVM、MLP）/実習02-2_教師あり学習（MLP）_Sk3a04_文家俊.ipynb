# 教師あり学習（多層パーセプトロン、MLP）
---
今回は、教師あり学習のアルゴリズムである多層パーセプトロン（MLP）を取り扱う。
MLPを用いたクラス分類（MLPClassifier）では、ノードや層の数を増やすことで、複雑な分割が行えるようになる。

[一部参考：sklearn.neural_network.MLPClassifier.html](https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html)

基本的な手順は以下である。
1. データセットの準備
2. モデルを決める
3. モデルを訓練する
4. モデルを評価する
5. モデルを使用する（新しい入力値に対する予測）

## 1. 基本的な動作の確認：XORの判定
### 1-1. データセットの準備
簡単のため、XORを想定し、
2つの入力に対して、クラス０かクラス１に分類する問題を考える。
次の教師データを用意する。
* [0,0]を入力するとクラス0を出力
* [0,1]を入力するとクラス1を出力
* [1,0]を入力するとクラス1を出力
* [1,1]を入力するとクラス0を出力

入力と出力を、それぞれ変数X, yに代入しておく。
```
# 教師データの入力X、出力y
X = [[0,0],[0,1],[1,0],[1,1]]
y = [0,1,1,0]
```

**今回の学習分類は、教師あり学習（分類）です。**

````
# 教師データの入力X、出力y
X = [[0,0],[0,1],[1,0],[1,1]]
y = [0,1,1,0]
````

## 2. アヤメの分類
### 2-1. データセットの準備
ここでは、sklearnに用意されているアヤメのデータセットを使用した分類を行ってみる。手順は前回の実習も参考にすること。

データセットの詳細：[sklearn.datasets.load_iris](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html)

```
# データセットを読み込み、変数に入れる。
from sklearn.datasets import load_iris
dataset = load_iris()
dataset #表示して確認
```

````
# データセットを読み込み、変数に入れる。
from sklearn.datasets import load_iris
dataset = load_iris()
dataset
````

datasetのdata(入力)とtarget（出力となる分類番号）を変数（xとtなど）に入れておく。

````
# 変数に入れておく
x = dataset.data
t = dataset.target

print("data: ",x)
print("target: ",t)
````

データセットを、訓練用とテスト用に分割する。
```
# データセットを分割する関数の読み込み
from sklearn.model_selection import train_test_split

# 訓練用データセットとテスト用データセットへの分割（訓練用80%、テスト用20%、random_stateは学籍番号）
x_train, x_test, t_train, t_test = train_test_split(x, t, test_size=0.2, random_state=学籍番号)
```

````
# データセットを分割する関数の読み込み
from sklearn.model_selection import train_test_split
# 訓練用データセットとテスト用データセットへの分割（訓練用80%、テスト用20%）
x_train, x_test, t_train, t_test = train_test_split(x, t, test_size=0.2, random_state=2220042)
````

### 2-2. モデルを決める、2-3. モデルを訓練する
MLPのモデルを作成し、
fit()関数に訓練データを入力してモデルを訓練する

````
#モデル作成と訓練
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(max_iter=2000)
clf.fit(x_train, t_train)
````

### 2-4. モデルを評価する
**【チェック】score()にテスト用データを入力し、正解率を求める。**

````
# テスト用データの正解率を求める
print(clf.score(x_test, t_test))
````

参考：訓練データに対する正解率を求める。

````
# 参考：訓練データに対する正解率を求める
clf.score(x_train, t_train)
````

### 2-5. モデルを使用する
テストデータの最初のデータなど入力して予測を出してみよう。

```
# 直接入力する例
clf.predict([[6.4, 2.8, 5.6, 2.2]])
# スライスを使用してリストの一部を入力する例
clf.predict(x_test[:1]) # 0番目
clf.predict(x_test[5:10]) # 5番目から9番目
```
**print文を使用しない場合は最後しか表示されないので注意**

````
# 予測（予測と正解を両方表示すると分かりやすい）
clf.predict([[6.4, 2.8, 5.6, 2.2]])
print("予測値:", clf.predict(x_test[:1]))
print("実際の値:", t_test[0])
````
