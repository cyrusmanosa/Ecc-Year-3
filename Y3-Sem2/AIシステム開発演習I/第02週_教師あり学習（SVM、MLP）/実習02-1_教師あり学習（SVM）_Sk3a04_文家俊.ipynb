# 教師あり学習（サポートベクターマシン、SVM）
---
今回は、教師あり学習のアルゴリズムであるサポートベクターマシン（SVM）を取り扱う。

## 1. 基本的な動作の確認：XORの判定
### 1-1. データセットの準備
簡単のため、XORを想定し、
2つの入力に対して、クラス０かクラス１に分類する問題を考える。
次の教師データを用意する。
* [0,0]を入力するとクラス0を出力
* [0,1]を入力するとクラス1を出力
* [1,0]を入力するとクラス1を出力
* [1,1]を入力するとクラス0を出力

入力と出力を、それぞれ変数X, yに代入しておく。
```
# 教師データの入力X、出力y
X = [[0,0],[0,1],[1,0],[1,1]]
y = [0,1,1,0]
```

**今回の学習分類は、教師あり学習（分類）です。**

````
# 教師データの入力X、出力y
X = [[0,0],[0,1],[1,0],[1,1]]
y = [0,1,1,0]
````

ロジスティック回帰を使用した場合(XOR)
---


### （ロジスティック回帰）1-2 モデルを決める、1-3 モデルを訓練する
分類問題のアルゴリズムからロジスティック回帰を選択し、モデルを用意する。さらに、.fit()の引数として入力Xと出力yを与えて学習させる。

```
# ロジスティック回帰を使用するためインポートする
from sklearn.linear_model import LogisticRegression
# モデルを用意する
clf_LR = LogisticRegression()
# 教師データX, yを使用して学習する
clf_LR.fit(X, y)
```

**ロジスティック回帰は分類問題用のアルゴリズム。間違いやすいので注意！**

````
#インポート
from sklearn.linear_model import LogisticRegression
#モデルの準備
clf_LR = LogisticRegression()
#学習
clf_LR.fit(X, y)
````

### （ロジスティック回帰）1-4 モデルを評価する、1-5 モデルを使用する
.score()やpredict()を使用して、正しく分類できるかを確認する。
*   .score(入力, 出力)は分類問題の場合は正解率を表す（1は全て正解）
*   .predict()について
  *   引数として、[[0,0],[1,0]]など、複数の入力を一度に計算可能
  *   入力は整数でなくても可（[0.2,0.7]など）
  *   1つの入力でも、[[0,0]]のようにリストのリストで書く

```
# .score()で正しく分類できているか確認
print(clf_LR.score(X,y))
# .predict()により正しく分類できるか確認
clf_LR.predict([[5,5],[1,0],[0.2,0.5],[2,2]])
```


**【チェック】正しく分類できていますか？（X, yを入れた場合のscoreの結果）**

**[ できていない ]**

````
# .score()で正しく分類できているか確認
print("正解率：",clf_LR.score(X,y))
# .predict()により正しく分類できるか確認
clf_LR.predict([[5,5],[1,0],[0.2,0.5],[2,2]])
# clf_LR.predict(X)
````

## 2. アヤメの分類
### 2-1. データセットの準備
ここでは、sklearnに用意されているアヤメのデータセットを使用した分類を行ってみる。手順は前回の実習も参考にすること。

データセットの詳細：[sklearn.datasets.load_iris](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html)

```
# データセットを読み込み、変数に入れる。

dataset #表示して確認
```

**【チェック】入力データ1つの次元（1つのアヤメを表す数値の数）と出力の分類数（何種類に分類するか）はいくつですか。<br>（上のリンク先か下のdatasetのdataとtargetから調べる）**
*   **入力データ1つの次元：[   4   ]**
*   **分類数：[   3   ]**


````
# データセットを読み込み、変数に入れる。
from sklearn.datasets import load_iris
dataset = load_iris()
print(dataset)
````

datasetのdata(入力)とtarget（出力となる分類番号）を変数（xとtなど）に入れておく。

````
# 変数に入れておく
t = dataset.target
x = dataset.data

# Check Data
t
````

データセットを、訓練用とテスト用に分割する。
```
# データセットを分割する関数の読み込み
from sklearn.model_selection import train_test_split

# 訓練用データセットとテスト用データセットへの分割（訓練用80%、テスト用20%）
x_train, x_test, t_train, t_test = train_test_split(x, t, test_size=0.2)
```

````
# データセットを分割する関数の読み込み
from sklearn.model_selection import train_test_split
# 訓練用データセットとテスト用データセットへの分割（訓練用80%、テスト用20%）
x_train, x_test, t_train, t_test = train_test_split(x, t, test_size=0.2)
````

### 2-2. モデルを決める、2-3. モデルを訓練する
SVCのモデルを作成し、
fit()関数に訓練データを入力してモデルを訓練する

````
#モデル作成と訓練
model = SVC()
model.fit(x_train, t_train)
````

### 2-4. モデルを評価する
**【チェック】score()にテスト用データを入力し、正解率を求める。**

````
# テスト用データの正解率を求める
model.score(x_test, t_test)
````

参考：訓練データに対する正解率を求める。

````
# 参考：訓練データに対する正解率を求める
model.score(x_train, t_train)
````

### 2-5. モデルを使用する
テストデータの最初のデータなど入力して予測を出してみよう。

```
# 直接入力する例
clf.predict([[6.4, 2.8, 5.6, 2.2]])
# スライスを使用してリストの一部を入力する例
clf.predict(x_test[:1]) # 0番目
clf.predict(x_test[5:10]) # 5番目から9番目
```
**print文を使用しない場合は最後しか表示されないので注意**

````
# 予測（予測と正解を両方表示すると分かりやすい）
print("予測値:", model.predict(x_test[:1]))
print("実際の値:", t_test[0])
````
